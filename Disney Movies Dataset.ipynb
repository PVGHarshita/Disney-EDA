# Which 2 months typically have the highest releases in a year? 

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sept","Oct","Nov","Dec"]
release_month = []
for i in range(df.shape[0]):
    release_month.append(months[int(df["release_date"][i][5:7])-1])
df["Release_Month"]=release_month
df.head()

#here use the inbuilt functions to convert date to month or day
```

```python
df["Release_Month"].value_counts()
```

# Movies released in which 2 months have the highest average gross collection? 

```python
avg__month_gross=[]
for i in months:
    avg__month_gross.append(df[df["Release_Month"]==i]["total_gross"].mean())
    
# months1=months[::]
# for i in range(len(months1)):
#     max_ind=i
#     for j in range(i+1,len(months1)):
#         if avg__month_gross[j]>avg__month_gross[max_ind]:
#             max_ind=j
#     avg__month_gross[i],avg__month_gross[max_ind]=avg__month_gross[max_ind],avg__month_gross[i]
#     months1[i],months1[max_ind]=months1[max_ind],months1[i]

# months1

df1 = pd.DataFrame()
df1["Months"]=months
df1["Monthly_gross"]=avg__month_gross
df1.sort_values(by=["Monthly_gross"])
```

# Which day of the week has the highest no. of movie releases & how many movies have been released until now on that day? 

```python
days = {"Monday":0,"Tuesday":0,"Wednesday":0,"Thursday":0,"Friday":0,"Saturday":0,"Sunday":0}
for i in range(df.shape[0]):
    k = df["release_date"][i][8:10]+" "+df["release_date"][i][5:7]+" "+df["release_date"][i][:4]
    born = datetime.datetime.strptime(k, "%d %m %Y").weekday()
    days[calendar.day_name[born]]+=1
days

#use the inbuilt function to convert the date to day
```
